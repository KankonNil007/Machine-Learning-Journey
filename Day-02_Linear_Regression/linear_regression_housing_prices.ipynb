```python
X = df.drop(['Price', 'Address', 'City'], axis = 1)
y = df['Price']
X
```

```python
X_encoded_1 = pd.get_dummies(X, columns = ['County','State'], drop_first=True)
X_encoded_1 = X_encoded_1.astype(int)
X_encoded_1.head()
```

```python
# Chi - Squared Test
from scipy.stats import chi2_contingency

cat_features = ['County_Alameda', 'County_Anne Arundel', 'County_Anoka', 'County_Arapahoe', 'County_Atascosa', 'County_Baltimore', 'County_Bernalillo', 'County_Bexar', 'County_Bronx', 'County_Bullitt', 'County_Canadian', 'County_Chesapeake', 'County_Clark', 'County_Clay', 'County_Cleveland', 'County_Cobb', 'County_Collin', 'County_Cook', 'County_Creek', 'County_Dallas', 'County_Davidson', 'County_DeKalb', 'County_Delaware', 'County_Denton', 'County_Denver', 'County_District of Columbia', 'County_Douglas', 'County_Durham', 'County_Duval', 'County_El Paso', 'County_Fairfield', 'County_Fort Bend', 'County_Franklin', 'County_Fresno', 'County_Fulton', 'County_Hamilton', 'County_Hancock', 'County_Harris', 'County_Hays', 'County_Hendricks', 'County_Hennepin', 'County_Jackson', 'County_Jefferson', 'County_Johnson', 'County_Kendall', 'County_King', 'County_Kings', 'County_Los Angeles', 'County_Madera', 'County_Maricopa', 'County_Marion', 'County_Mecklenburg', 'County_Miami-Dade', 'County_Milwaukee', 'County_Multnomah', 'County_Nassau', 'County_New York', 'County_Oklahoma', 'County_Oldham', 'County_Orleans', 'County_Osage', 'County_Parker', 'County_Philadelphia', 'County_Pima', 'County_Pinal', 'County_Platte', 'County_Queens', 'County_Richmond', 'County_Sacramento', 'County_San Diego', 'County_San Francisco', 'County_Santa Clara', 'County_Sarpy', 'County_Sedgwick', 'County_Shelby', 'County_St. Johns', 'County_Tarrant', 'County_Travis', 'County_Tulsa', 'County_Union', 'County_Virginia Beach', 'County_Wagoner', 'County_Wake', 'County_Washington', 'County_Wayne', 'County_Williamson', 'County_Wyandotte', 'State_California', 'State_Colorado', 'State_District of Columbia', 'State_Florida', 'State_Georgia', 'State_Illinois', 'State_Indiana', 'State_Kansas', 'State_Kentucky', 'State_Louisiana', 'State_Maryland', 'State_Michigan', 'State_Minnesota', 'State_Missouri', 'State_Nebraska', 'State_Nevada', 'State_New Mexico', 'State_New York', 'State_North Carolina', 'State_Ohio', 'State_Oklahoma', 'State_Oregon', 'State_Pennsylvania', 'State_Tennessee', 'State_Texas', 'State_Virginia', 'State_Washington', 'State_Wisconsin']
alpha = 0.05
chi2_results = {}

X_encoded_1['charges_bin'] = pd.qcut(df['Price'], q=4, labels=False)

for col in cat_features:
    contingency = pd.crosstab(X_encoded_1[col], X_encoded_1['charges_bin'])
    chi2_stat, p_val, _, _ = chi2_contingency(contingency)
    decision = 'Reject Null (Keep Feature)' if p_val < alpha else 'Accept Null (Drop Feature)'
    chi2_results[col] = {
        'chi2_statistic': chi2_stat,
        'p_value': p_val,
        'decision': decision
    }

chi2_df = pd.DataFrame(chi2_results).T
chi2_df = chi2_df.sort_values(by='p_value')
with pd.option_context('display.max_rows', None):
    display(chi2_df)
```

```python
X_encoded_1.drop(columns=['County_Delaware', 'County_Anne Arundel', 'County_Cobb',
                                        'County_Bronx', 'County_Anoka', 'County_Atascosa',
                                        'County_Kendall', 'County_Arapahoe', 'County_Durham', 'County_Union',
                                        'County_Nassau', 'County_Bullitt', 'County_Wagoner',
                                        'County_St. Johns', 'County_Oldham', 'County_Chesapeake', 'County_Creek'], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

cols = ['County','State']

X_label = X.copy()

for i in cols:
    X_label[i] = encoder.fit_transform(X_label[i])

X_label
```

```python
from sklearn.preprocessing import StandardScaler

cols = ['Zip Code', 'Beds', 'Baths', 'Living Space', 'Zip Code Population', 'Zip Code Density',
       'Median Household Income', 'Latitude', 'Longitude']
scaler = StandardScaler()

X_encoded_1[cols] = scaler.fit_transform(X_encoded_1[cols])
X_encoded_1
```

```python
cols2 = ['Zip Code', 'Beds', 'Baths', 'Living Space', 'Zip Code Population', 'Zip Code Density', 'County',
       'Median Household Income', 'Latitude', 'Longitude','State']
scaler2 = StandardScaler()

X_label[cols2] = scaler2.fit_transform(X_label[cols2])
X_label
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded_1, y, test_size=0.20, random_state=42)
model1 = LinearRegression()
model1.fit(X_train, y_train)
```

```python
y_pred = model1.predict(X_test)
r2 = r2_score(y_test, y_pred)
r2
```

```python
n = X_test.shape[0]
p = X_test.shape[1]

adjusted_r2 = 1 - ((1 - r2) * (n - 1) / (n - p - 1))

adjusted_r2
```
