```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
X, y = make_blobs(n_samples=500, n_features=5, centers=3, cluster_std=0.5, random_state=42)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
df = pd.DataFrame(X_pca, columns=["Feature1", "Feature2"])
df['results'] = y
```

```python
sns.scatterplot(data=df, x="Feature1", y="Feature2", hue="results", palette="Set2")
```
